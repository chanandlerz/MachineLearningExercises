### *Deteksi dan Penanganan Outlier*

```python
numerical_data = data.select_dtypes(include=[np.number])
```

```python
Q1 = numerical_data.quantile(0.25)
Q3 = numerical_data.quantile(0.75)
IQR = Q3 - Q1

no_outlier_df = data[~((numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

# **6. Pembangunan Model Clustering**

## **a. Pembangunan Model Clustering**

Pada tahap ini, Anda membangun model clustering dengan memilih algoritma yang sesuai untuk mengelompokkan data berdasarkan kesamaan. Berikut adalah **rekomendasi** tahapannya.
1. Pilih algoritma clustering yang sesuai.
2. Latih model dengan data menggunakan algoritma tersebut.

```python
cleaned_data = no_invoiceDate_df.copy()
```

```python
X = cleaned_data.iloc[:,[3,4]]

X
```

```python
kmeans = KMeans()

visualizer = KElbowVisualizer(kmeans, k=(1,10))

visualizer.fit(X)

visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_

k = 3
```

```python
X
```

```python
def analyze_cluster(X, labels, k):
    print("Analisis Karakteristik Setiap Cluster: ")

    for cluster_id in range(k):
        cluster_data = X[labels == cluster_id]

        mean_quantity = cluster_data['Quantity'].mean()
        mean_unit_price = cluster_data['UnitPrice'].mean()


        print(f"Rata-rata quantity cluster_{cluster_id} = {mean_quantity:.2f}")
        print(f"Rata-rata bulan cluster_{cluster_id} = {mean_unit_price:.2f}")

analyze_cluster(X, labels, k)
```

```python
print(labels)
```

```python
X
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.figure(figsize=(12,8))

plt.scatter(X['Quantity'], X['UnitPrice'],c=labels, cmap='viridis', s=50, alpha=0.6, edgecolors='w', marker='o')

plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')

print(centroids)
for i, centroid in enumerate(centroids):
    # print(centroid)
    plt.text(centroid[0], centroid[1], f'Centroid {i+1}', color='red', fontsize=12, ha='center', va='center')

plt.title('Visualisasi Cluster dengan Centroid')
plt.xlabel('Quantity')
plt.ylabel('Unit Price')
plt.legend()
 
plt.show()


print("Nilai Centroids:")
for i, centroid in enumerate(centroids):
    print(f"Centroid {i+1}: Quantity = {centroid[0]:.2f}, Unit Price = {centroid[1]:.2f}")
```
